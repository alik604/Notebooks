## Ensembling with VotingClassifier
https://stats.stackexchange.com/questions/190151/ensembling-with-votingclassifier
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingClassifier.html#sklearn.ensemble.VotingClassifier

```python
def benchmark_mustPrimeWithTrain_Test_BeforeHand():
    topI = 0
    topY = 0
    clf_KnC = KNeighborsClassifier(5)
    for i in range(1,40,1):
        clf = KNeighborsClassifier(i)
        clfF = clf.fit(X_train,y_train)
        y_pred = clfF.predict(X_test)
        score = accuracy_score(y_test, y_pred)
        if score > topY:
            topY=score
            topI = i
            clf_KnC=clfF

    #     print("i: {0}, Accuracy: {1}".format(i,score))
    print("BEST: i: {0}, Accuracy: {1}".format(topI,topY))


    alg1 = RandomForestClassifier(criterion='entropy',max_depth=5) #ExtraTreesClassifier()
    alg2 = GradientBoostingClassifier()
    alg3 = clf_KnC #KNeighborsClassifier(topI)

    eAlg = VotingClassifier(estimators=[('RandomForestClassifier',alg1),('GradientBoostingClassifier',alg2),('KNeighborsClassifier',alg3)],
                          voting='hard', weights=[1,1,1]).fit(X_train,y_train)

    #eAlg.fit(X_train,y_train)
    pred_ens = eAlg.predict(X_test)

    MSE= mean_squared_error(pred_ens,y_test)  
    MAD= mean_absolute_error(pred_ens,y_test)  
    ac = accuracy_score(pred_ens, y_test)

    print("MSE:", MSE)
    print("MAD:", MAD)
    print("AC:", ac)
    
benchmark_mustPrimeWithTrain_Test_BeforeHand()
```

```python
# Feature Importance with Extra Trees Classifier

def getFeatureImportance(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.99, random_state=1) # 70% training and 30% test

    model = ExtraTreesClassifier(n_estimators=10)#,random_state=1
    model.fit(X_train, y_train)
    return model.feature_importances_

featureImportances_base= getFeatureImportance(X,y)

for i in range(10): 
    featureImportances_base = featureImportances_base + getFeatureImportance(X,y)

#featureImportances_base   
l = dict(list(zip(X_train.columns, featureImportances_base)))

l.pop(min(l, key=l.get))
l.pop(min(l, key=l.get))

keepCol=list(l.keys())
keepCol
```
